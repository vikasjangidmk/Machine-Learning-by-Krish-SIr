```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
# Independent Features
X = pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
X.head()
```

```python
# Dependent Features
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
## Apply Decision Tree Classifier 
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
## Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cmd = ConfusionMatrixDisplay(confusion_matrix=cm)

# Plot the confusion matrix
cmd.plot(cmap=plt.cm.Blues)  # You can choose a different colormap if you want
plt.show()
```

```python
# Assuming y_test and y_pred are defined
report = classification_report(y_test, y_pred, output_dict=True)

# Print the classification report
print(classification_report(y_test, y_pred))

# Convert the report into a DataFrame for easier plotting
report_df = pd.DataFrame(report).transpose()

# Plot the classification report
plt.figure(figsize=(10, 6))
sns.heatmap(report_df.iloc[:-1, :].T, annot=True, cmap='Blues', fmt='.2f')
plt.title('Classification Report')
plt.xlabel('Classes')
plt.ylabel('Metrics')
plt.show()
```

---

**Decision Tree Classifier Prepruning And Hyperparameters Tunning**

| Feature             | Pre-Pruning                          | Post-Pruning                          |
|---------------------|-------------------------------------|--------------------------------------|
| **Definition**      | Stops tree growth during training.  | Prunes the tree after it is fully grown. |
| **Method**          | Evaluates splits and stops if improvement is insufficient. | Evaluates the full tree and removes less important nodes. |
| **Advantages**      | - Reduces computational cost. <br> - Prevents overfitting early on. | - Captures all potential relationships initially. <br> - More flexible in adjusting complexity. |
| **Disadvantages**   | - May lead to underfitting if criteria are too strict. | - Computationally expensive. <br> - Risk of overfitting if pruning is not done carefully. |
| **Parameters**      | - `min_samples_split`: Minimum samples required to split a node. <br> - `min_samples_leaf`: Minimum samples required at a leaf node. <br> - `max_depth`: Maximum depth of the tree. | - `ccp_alpha`: Complexity parameter used for pruning. <br> - Cross-validation scores to evaluate the effectiveness of pruning. |
| **Use Case**        | Best when simplicity is prioritized. | Useful when you want to capture complex patterns before simplifying. |
